```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies.csv")
```

```python
df["success"] = (df["revenue"] > df["budget"]).astype(int)
```

```python
success = df[df["success"] == 1]["popularity"]

failure = df[df["success"] == 0]["popularity"]
```

```python
t_stat, p_value = ttest_ind(success, failure)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("Popularity differs significantly between successful and unsuccessful movies.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant difference in popularity.")
```

# Chi-Square Test

```python
from scipy.stats import chi2_contingency
```

```python
table = pd.crosstab(df["genres"], df["success"])

print(table)
```

```python
chi2, p, dof, expected = chi2_contingency(table)

print("Chi-Square:", chi2)
print("P Value:", p)
```

```python
if p < 0.05:
    print("Reject Null Hypothesis")
    print("Genre and movie success are associated.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant association between genre and movie success.")
```
